## Data Type Conversion

Some columns contained data in string format which should be numeric (like `Price`, `Rating`, etc.). These were converted to appropriate numeric types for accurate analysis.


```python
  # Remove non-numeric characters from discounted_price and actual_price
data['discounted_price'] = data['discounted_price'].str.replace(r'[^\d.]', '', regex=True)
data['actual_price'] = data['actual_price'].str.replace(r'[^\d.]', '', regex=True)

# Replace empty strings with NaN, then convert to float
data['discounted_price'] = pd.to_numeric(data['discounted_price'], errors='coerce')
data['actual_price'] = pd.to_numeric(data['actual_price'], errors='coerce')

# Convert discount_percentage (remove % and convert to float)
data['discount_percentage'] = data['discount_percentage'].str.replace('%', '', regex=False).astype(float)

# Convert rating to float
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')

# Convert rating_count to int (remove commas first)
data['rating_count'] = data['rating_count'].str.replace(',', '', regex=True).astype(float).astype('Int64')

# Check if there are any NaN values after conversion
print(data[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].isna().sum())

# Display updated data types
print(data.dtypes)
```

## Simplifying Top 10 Category Names 

The original category names were too long and made visualizations difficult to read. To improve clarity, only the first and last parts of each category path were retained.


```python
data['Reduced_Category'] = data['category'].apply(lambda x: '|'.join([x.split('|')[0], x.split('|')[-1]]))

#Calculate average rating for each reduced category
category_ratings = data.groupby('Reduced_Category')['rating'].mean().reset_index()

# Get top 10 categories by average rating
top_10_categories = category_ratings.sort_values(by='rating', ascending=False).head(10)

#Filter the original data to keep only top 10 rated reduced categories
top_category_names = top_10_categories['Reduced_Category']
data_top10 = data[data['Reduced_Category'].isin(top_category_names)]
```

# Visualization 


## Discount & Pricing Analysis

### Analyzing Discount Trends 

```python
plt.figure(figsize = (8,5))
sns.histplot(data['discount_percentage'], bins = 20, kde = True)
plt.title("Distribution of Discounts")
```

The majority of discounts on Amazon fall between 50-60%, indicating that sellers prefer moderate discounting over extreme price cuts. Products with very high (80%+) or very low (10%-) discounts are less frequent, suggesting that moderate discounts might be the most effective in attracting buyers.

### Catergory- wise Discount Percentage


```python
# Group by 'category' and calculate the average discount percentage
category_discount = data.groupby("Reduced_Category")["discount_percentage"].mean().reset_index()
```

```python
#sort categories by discount percentage 
category_discount = category_discount.sort_values(by="discount_percentage", ascending = False).head(10)
```

```python
plt.figure(figsize = (8,5))
sns.barplot(data= category_discount, x = "discount_percentage", y = "Reduced_Category", palette ="viridis")

plt.xlabel("Averge Discount Percentage")
plt.ylabel("Category")
plt.title("Average Discount Percentage by top 10 category")
plt.show()
```

High discounts are heavily concentrated in electronics-related accessories.
Categories such as Cable Connection Protectors, Earpads, and Phone Charms have average discounts exceeding 75%. This indicates a highly competitive or oversaturated segment where sellers may rely on steep discounts to boost visibility and sales.

## Customer and Sales Analysis 

### Rating vs Discount percentage 

```python
plt.figure(figsize = (10,6))
sns.scatterplot(data = data, x = 'discount_percentage', y = 'rating', alpha = 0.5)

sns.regplot(data = data, x = 'discount_percentage', y = 'rating', scatter = False, color = 'red')

plt.xlabel("Discount Percentage")
plt.ylabel("Rating")
plt.title("Relationship Between Discount Percentage and Rating")
plt.grid(True)
plt.show()
```

A higher discount percentage may slightly reduce product ratings.

### Product Price vs. Discount vs. Rating 


```python
# Select top 10 categories based on average rating
top_categories = data.groupby('Reduced_Category')['rating'].mean().nlargest(10).index
top_data = data[data['Reduced_Category'].isin(top_categories)]

# Create a pivot table: rows = Reduced_Category, columns = rating, values = average discount
heatmap_data = top_data.pivot_table(
    index='Reduced_Category',
    columns='rating',
    values='discount_percentage',
    aggfunc='mean'
)

# Plot the heatmap
plt.figure(figsize=(15, 4))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Average Discount Percentage by Rating and Category')
plt.xlabel('Rating')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

High-rated categories like Cord Management, Memory, and Power LAN Adapters receive significant discounts, suggesting competitive pricing or stock clearance strategies. In contrast, categories like Painting Materials and Streaming Clients maintain high ratings with minimal or no discounts.
